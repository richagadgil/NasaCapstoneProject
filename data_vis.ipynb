```
# !pip install plotly --upgrade

import plotly.express as px
import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```
spectra = loadmat('/content/drive/MyDrive/Spectra_3s_25c.mat')
spectra.keys()
```

# Junk Testing

```
mean_spectra=[]
spectra_number=0
for x in spectra['Dm_total_save']:
    mean_val=0
    for y in x:
        mean_val += y[spectra_number]
    mean_spectra.append(mean_val/len(x))
```

```
modes = [x[spectra_number] for x in spectra['omega_total_save']]
t_modes = np.sort([x*1000000 for x in modes])
```

```
peaks, properties = find_peaks(mean_spectra, prominence=0.01)
```

```
len(peaks)
```

```
temp_modes = pd.Series(t_modes)[(pd.Series(t_modes) < 991000)]
temp_modes = pd.Series(temp_modes)[(pd.Series(temp_modes) > 990000)]
temp_peaks = pd.Series(peaks)[(pd.Series(peaks) < 991000)]
temp_peaks = pd.Series(temp_peaks)[(pd.Series(temp_peaks) > 990000)]
```

```
fig = px.line(x = range(990000,991000),y= mean_spectra[990000:991000], width=1000, height=400)
for mode in temp_modes:
    fig.add_vline(x=mode, line_width=1, line_color="green") #requires plotly 4.12 and above
for peak in temp_peaks:
    fig.add_vline(x=peak, line_width=1, line_dash="dash", line_color="red") #requires plotly 4.12 and above
fig.show()
```

```
fig = px.line(x=range(1000000-10000, 1000000),y=mean_spectra[-10000:], width=1000, height=400)
for mode in t_modes[-200:]:
    fig.add_vline(x=mode, line_width=1, line_color="green") #requires plotly 4.12 and above
for peak in peaks[-200:]:
    fig.add_vline(x=peak, line_width=1, line_dash="dash", line_color="red") #requires plotly 4.12 and above
fig.show()
```

```
pd.options.plotting.backend = "plotly" # set default pandas plot to plotly
```

```
spectra_number=0
```

```
all_channels = pd.DataFrame(np.transpose(np.transpose(spectra['Dm_total_save'])[spectra_number]))
all_channels.head()
```

```
all_peaks=[]
for c in all_channels:
    peaks, properties = find_peaks(all_channels[c], prominence=0.5)
    all_peaks.extend(peaks)
all_peaks=np.sort(all_peaks)
```

```
# don't make the range too wide or it will take forever to run
view_range = (990000, 991000)
```

```
modes = [x[spectra_number] for x in spectra['omega_total_save']]
t_modes = np.sort([x*1000000 for x in modes])
temp_modes = pd.Series(t_modes)[(pd.Series(t_modes) < view_range[1]]
temp_modes = pd.Series(temp_modes)[(pd.Series(temp_modes) > view_range[0])]
temp_peaks = pd.Series(all_peaks)[(pd.Series(all_peaks) < view_range[1])]
temp_peaks = pd.Series(temp_peaks)[(pd.Series(temp_peaks) > view_range[0])]
```

```
temp_peaks = list(pd.Series(temp_peaks).unique())
```

```
fig = all_channels.iloc[range(view_range[0],view_range[1])].plot(template='plotly_dark', kind='line')
for mode in temp_modes:
    fig.add_vline(x=mode, line_width=1, line_color="green") #requires plotly 4.12 and above
for peak in temp_peaks:
    fig.add_vline(x=peak, line_width=1, line_dash="dash", line_color="red") #requires plotly 4.12 and above
fig.show()
```
